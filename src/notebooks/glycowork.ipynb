```python
# Imports
import pandas as pd
import glycowork as gw 
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load glycan data
from glycowork.glycan_data.loader import glycan_binding, df_glycan, df_species
```

```python
df_glycan.head()
```

```python
# Define the structures we're looking for
structures = {
    'Complex_Gal': "Gal(b1-4)GlcNAc(b1-2)Man(a1-6)[Gal(b1-4)GlcNAc(b1-2)Man(a1-3)]Man(b1-4)GlcNAc(b1-4)GlcNAc",
    'Complex_Fuc': "GlcNAc(b1-2)Man(a1-6)[GlcNAc(b1-2)Man(a1-3)]Man(b1-4)GlcNAc(b1-4)[Fuc(a1-6)]GlcNAc", 
    'Complex_Sia': "NeuAc(a2-6)Gal(b1-4)GlcNAc(b1-2)Man(a1-6)[NeuAc(a2-6)Gal(b1-4)GlcNAc(b1-2)Man(a1-3)]Man(b1-4)GlcNAc(b1-4)GlcNAc"
}

# Create masks for each structure
masks = {name: df_glycan['glycan'].str.contains(struct, regex=False, na=False) 
         for name, struct in structures.items()}

# Combine all matches into one DataFrame
N_glycans_df = pd.concat([
    df_glycan[mask].assign(Structure_Type=name)
    for name, mask in masks.items()
], ignore_index=True)
```
